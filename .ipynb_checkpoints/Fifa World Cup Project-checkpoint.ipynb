```python
# Dataframe overview function
def dataframe_overview(df):
    print('Dataframe Info:\n')
    df.info()

    print('\n Dataframe Description:\n')
    print(df.describe())
    
    print("\n Check null value in Dataframe: \n")
    print(df.isnull().sum())

    print('\n Check Duplicate values in Datafrme: \n')
    print(df.duplicated().sum())
    
    print('\n Check the unique values of dataframe: \n')
    print(df.nunique())

    print('\n Column Description: \n')
    print(list(df.columns))

    print('\n Shape of Dataframe (Rows,Colums):\n')
    print(df.shape)

    print('\n Data type of Each Column in Dataframe: \n ')
    print(df.dtypes)

    return
```

```python
# Pass dataframes to overview
print('\n Displaying the Overview of worldcup matches  dataframe: \n')
dataframe_overview(matches)

```

```python
# Cleaning function
def trim(df):
    df.columns = df.columns.str.strip()
    df = df.drop_duplicates()
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')

    df_obj = df.select_dtypes(['object'])
    df[df_obj.columns] = df_obj.apply(lambda x: x.astype(str).str.strip())

    print("All column names cleaned, duplicates dropped, and text columns processed.")
    print("Clean dataset \n")
    print(df.head())
    return df
```

```python
# Apply cleaning function
print("\n Clean  Dataset")
matches_df = trim(matches)
print(" dataset cleaned.")

print("\n Clean  Dataset")
players_df = trim(players)
print(" dataset cleaned.")

print("\n Clean  Dataset")
world_cup_df = trim(cups)
print(" dataset cleaned.")
```

```python
#We have to import these software: matplotlib is for scientific research analysis and seaborn is for graphical illustration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Hypothesis 1: Attendance Over Time ---
# Assuming 'Year' and 'Attendance' columns exist in world_cup_df
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(cups, x='year', y='attendance', marker='o', color='blue')
plt.title("World Cup attendance Over Time")
plt.xlabel("year")
plt.ylabel("Total attendance")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Hypothesis 2: Goal Trend Over Time ---
# We’ll calculate total goals per year from match_df
# Assuming columns: 'Year', 'Home Goals', 'Away Goals' or 'Home Team Goals', 'Away Team Goals'
```

```python
match_df = pd.read_csv("Match_data_clean.csv")
```

```python
if 'Home Team Goals' in match_df.columns:
    match_df.rename(columns={
        'Home Team Goals': 'Home Goals',
        'Away Team Goals': 'Away Goals'
    }, inplace=True)
```

```python
match_df['Total Goals'] = match_df['home_team_goals'] + match_df['away_team_goals']
goals_per_year = match_df.groupby('year')['Total Goals'].sum().reset_index()
```

```python
# Plotting goal trend
plt.figure(figsize=(6, 4))
sns.lineplot(data=goals_per_year, x='year', y='Total Goals', marker='o', color='green')
plt.title("Total Goals Scored per World Cup Over Time")
plt.xlabel("year")
plt.ylabel("Total Goals")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
attendance_table = world_cup_df[['year', 'attendance']].sort_values(by='year')
```

```python
# Display the table
print(" attendance Over Time:")
print(attendance_table.to_string(index=False))
```

```python
# Calculate total goals per match
match_df['Total Goals'] = match_df['home_team_goals'] + match_df['away_team_goals']
```

```python
# Calculate number of matches and total goals per year
matches_per_year = match_df.groupby('year').size()
goals_per_year = match_df.groupby('year')['Total Goals'].sum()
```

```python
# Calculate average goals per match
avg_goals_per_match = (goals_per_year / matches_per_year).reset_index()
avg_goals_per_match.columns = ['year', 'Avg Goals per Match']
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(data=avg_goals_per_match, x='year', y='Avg Goals per Match', marker='o', color='darkred')
plt.title("Average Goals per Match Over Time in FIFA World Cup")
plt.xlabel("year")
plt.ylabel("Average Goals per Match")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
