```python
# Function to remove the � character from strings
def remove_bad_char(x):
    if isinstance(x, str):
        return x.replace('�', '')
    return x

df
```

```python
# Function to remove trailing zeros after decimal
def remove_trailing_zeros(x):
    if isinstance(x, float):
        return str(x).rstrip('0').rstrip('.') if '.' in str(x) else str(x)
    return x

df
```
